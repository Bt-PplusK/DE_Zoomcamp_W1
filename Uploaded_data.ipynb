```python
df_iter = pd.read_csv('green_tripdata_2019-10.csv.gz', iterator=True, chunksize=100000)
```

```python
df = next(df_iter)
```

```python
df
```

```python
# Find the maximum trip distance
max_lpep_pickup_datetime = df['tip_amount'].max()

# Filter rows where trip_distance is equal to the maximum
rows_with_lpep_pickup_datetime = df[df['tip_amount'] == max_lpep_pickup_datetime]

# Display the rows
print(rows_with_lpep_pickup_datetime)
```

```python
# Find the second maximum trip distance
second_max_trip_distance = df['lpep_pickup_datetime'].nlargest(2).iloc[-1]

# Filter rows where trip_distance is equal to the second maximum
rows_with_second_max_distance = df[df['lpep_pickup_datetime'] == second_max_trip_distance]

# Display the rows
print(rows_with_second_max_distance)
```

```python
from time import time
```

```python
while True: 
    t_start = time()

    df = next(df_iter)

    df.lpep_pickup_datetime = pd.to_datetime(df.lpep_pickup_datetime)
    df.lpep_dropoff_datetime = pd.to_datetime(df.lpep_dropoff_datetime)
    
    df.to_sql(name='green_taxi_data', con=engine, if_exists='append')

    t_end = time()

    print('inserted another chunk, took %.3f second' % (t_end - t_start))
```
